```julia
using ElectromagneticFields
```

```julia
using Makie
```

```julia
const B0 = 1.;
```

```julia
eq = SymmetricQuadratic(B0)
```

```julia
load_equilibrium(eq)
```

```julia
nr = 100
nz = 120
nl = 10
;
```

```julia
rgrid = LinRange(-2.0, +2.0, nr)
zgrid = LinRange(-2.0, +2.0, nz)
;
```

```julia
field = zeros((nr,nz))
potAR = zeros((nr,nz))
potAZ = zeros((nr,nz))
potAP = zeros((nr,nz))
;
```

```julia
for i in 1:nr
    for j in 1:nz
        field[i,j] = B(rgrid[i], zgrid[j], 0.0)
        potAR[i,j] = A₁(rgrid[i], zgrid[j], 0.0)
        potAZ[i,j] = A₂(rgrid[i], zgrid[j], 0.0)
        potAP[i,j] = A₃(rgrid[i], zgrid[j], 0.0)
    end
end
```

```julia
axis = (names = (axisnames = ("R", "Z"),),)
scene = vbox(
    contour(rgrid, zgrid, field, axis=axis, levels=nl, fillrange=true),#L"$B(R,Z)$"
    contour(rgrid, zgrid, potAR, axis=axis, levels=nl),#L"$A_R(R,Z)$"
    contour(rgrid, zgrid, potAZ, axis=axis, levels=nl) #L"$A_Z(R,Z)$"
)
```
